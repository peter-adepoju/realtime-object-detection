## App Architecture Overview

The Streamlit app (`app/streamlit_app.py`) is structured into four panels:

### Panel 1 — Sidebar: Configuration
- Model variant selector (nano / small / medium)
- Confidence threshold slider
- Class filter multi-select
- Video upload button

### Panel 2 — Live Detection
- Displays the annotated frame with bounding boxes
- Updates per-frame count overlay
- Shows current FPS

### Panel 3 — Count Dashboard
- Bar chart of current-frame detection counts
- Running total counts across all processed frames
- Time-series line chart of top 5 classes

### Panel 4 — Benchmark Metrics
- Table of pre-computed benchmark results
- Speed-accuracy trade-off scatter plot
- Model selector for comparison

````python
# ── Step 2: Check all app dependencies are available ──────────────────────────

dependencies = [
    ("streamlit", "Dashboard framework"),
    ("ultralytics", "YOLOv8 detection"),
    ("cv2", "Video frame processing"),
    ("plotly", "Interactive charts in Streamlit"),
    ("pandas", "Metrics tables"),
    ("numpy", "Array operations"),
]

print("App dependency check:")
all_ok = True
for module_name, purpose in dependencies:
    try:
        mod = __import__(module_name)
        version = getattr(mod, "__version__", "unknown")
        print(f"  ✓ {module_name:<14} v{version:<10}  — {purpose}")
    except ImportError:
        print(f"  ✗ {module_name:<14} NOT FOUND    — {purpose}")
        all_ok = False

if all_ok:
    print("\n✓ All app dependencies are available.")
else:
    print("\n⚠ Some dependencies are missing. Run: pip install -r requirements.txt")
````
